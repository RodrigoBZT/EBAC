#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na temperatura global, A NASA disponibiliza dados de diferenças de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. |ano|anomalia termica|

| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
series_anos = pd.Series(data=anos)
series_anos
```

```python
series_anomalia_termica = pd.Series(data=anomalia_termica)
series_anomalia_termica
```

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

```python
arr = np.random.normal(100, 10, (20,3))

# seu código aqui
df_arr = pd.DataFrame(data=arr)
df_arr
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#seu código aqui
indices = [i for i in range(1, 21)]
colunas = ['x1', 'x2', 'x3']
```

```python
df_arr = pd.DataFrame(data=arr, index=indices, columns=colunas)
df_arr
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui
df_arr['media'] = np.average(df_arr[['x1', 'x2', 'x3']], axis=1)
df_arr
```

```python
# seu código aqui
df_arr['log_med'] = np.log(df_arr['media'])
df_arr
```
